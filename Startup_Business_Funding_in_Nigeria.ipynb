```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("C:/Users/GABRIEL JOHNSON/Documents/Africas DeepTech Challage Prjoect/Dataset/Business Funding Data.csv", 
                      encoding='latin1')
dataset.head(21)
```

### Step1: Cleaning Dataset for Accurate Analysis

```python
# Replacing Null Values from 'Financing Type' Column
dataset['Financing Type'] = dataset['Financing Type'].fillna('Not Specified')
```

```python
dataset.head(20)
```

```python
dataset['Effective date'] = dataset['Effective date'].fillna('Till Further Notice')
```

```python
dataset.head(20)
```

```python
dataset['Investors'] = dataset['Investors'].fillna('Anonymous')
```

```python
dataset['Investors Count'] = dataset['Investors Count'].fillna(0)
```

```python
dataset['Financing Type Normalized'] = dataset['Financing Type Normalized'].fillna('Not Specified')
```

```python
dataset['Investors Count'] = dataset['Investors Count'].astype(int)
```

```python
dataset.head(20)
```

```python
dataset.dtypes
```

```python
dataset['Categories'] = dataset['Categories'].replace({'[]':'Not Specified'})
```

```python
dataset['Categories'] = dataset['Categories'].replace({'["private_equity"]':'Private Equity', 
                                                       '["debt_financing"]': "Debt Financing", 
                                                       '["seed", "venture"]': 'Seed Venture', 
                                                       '["series_i", "venture"]':'Series_i Venture', 
                                                      '["series_a2", "venture"]': 'Series_a2 Venture',
                                                      '["series_b", "venture"]':'Series_b Venture'})
```

```python
dataset.head(21)
```

```python
dataset[dataset.duplicated() == False]
```

```python
dataset['Categories'] = dataset['Categories'].replace({'["private_equity", "venture"]':'Private Equity', 
                                                       '["series_a", "venture"]': 'Series_a Venture'})
```

### Step2: Analyzing Dataset

```python
# Top Investors by Fund Ranking

investors = dataset.groupby("Investors").sum()

investors
```

```python
investors = investors.sort_values(by="Amount Normalized", ascending=False)
investors
```

```python
value = investors['Amount Normalized'].sort_values(ascending=False)
label = investors['Amount Normalized'].index
```

```python
plt.figure(figsize=(15, 5))
plt.bar(label, value, color='blue')
plt.ylabel("Fund Amount")
plt.xlabel("Investors Companies")
plt.xticks(rotation=90)
plt.title("Top Ranking Investors By Funding Amount")
plt.show()
```

```python
dataset.head(25)
```

```python
# Categorial Funding by Amount
categories = dataset.groupby('Categories').sum()
anon_sum = categories['Amount Normalized'].max()


categories = categories[(categories['Amount Normalized'] < categories['Amount Normalized'].max())]
categories
```

```python
x = categories['Amount Normalized']
y = categories['Amount Normalized'].index
```

```python
total_sum = x.sum()

total_sum
```

```python
anon_sum
```

```python
plt.pie(x, labels=y, colors=['blue', 'red', 'yellow', 'green', 'cyan', 'orange', 'purple'], autopct="%1.1f%%", shadow=True)
t = plt.title(f"Total Funding of ${total_sum:,} By Investment Type Categories \n(Excluding Anonymous Funding of ${anon_sum:,})\n")
plt.title(f"Total Funding of ${total_sum:,} By Investment Type Categories \n(Excluding Anonymous Funding of ${anon_sum:,})\n")
plt.show()
```

```python
t
```

```python
startups = dataset.groupby('Website Domain').sum()
startups = startups.sort_values(by='Amount Normalized', ascending=False)
startups
```

```python
v = startups['Amount Normalized']
l = startups['Amount Normalized'].index
```

```python
plt.figure(figsize=(15, 5))
plt.bar(l, v, color='green')
plt.xlabel("Fund Amount")
plt.ylabel("Start-Ups Companies (Millions $)")
plt.xticks(rotation=45)
plt.title("Top Ranking Start-Ups Companies By Funding Amount")
plt.show()
```
